```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras import regularizers
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
df = pd.read_csv('nation_one_obs.csv')
df.fillna(value=0, inplace=True)
data = df.to_numpy()[:,1:]
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
```

```python
input_dim = data.shape[1]  # 8
encoding_dim = 5
# Number of neurons in each Layer [8, 6, 4, 3, ...] of encoders
input_layer = Input(shape=(input_dim, ))
encoder_layer_1 = Dense(30, activation="tanh", activity_regularizer=regularizers.l1(10e-5))(input_layer)
bottleneck = Dense(encoding_dim, activation="tanh")(encoder_layer_1)
```

```python
# Create encoder model
encoder = Model(inputs=input_layer, outputs=bottleneck)
# Use the model to predict the factors which sum up the information of interest rates.
encoded_data = pd.DataFrame(encoder.predict(data_scaled))
encoded_data.columns = ['feat_'+ str(i) for i in range(encoding_dim)]
```
