```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nation_one_obs.csv')
df.fillna(value=0, inplace=True)
data = df.to_numpy()[:,1:]
```

```python
mean = data.mean(axis = 0)
data = (data - mean).astype(float)
stddev = np.std(data.astype(float),axis = 0)
Z = data / stddev
cov = Z.T @ Z
eigenValues, eigenVectors = np.linalg.eig(cov)
idx = eigenValues.argsort()[::-1]   
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:,idx]
```

```python
threshold = 0.8
sum = 0
totalSum = np.array([abs(v) for v in eigenValues]).sum()
for dimensions, v in enumerate(eigenValues):
    sum += abs(v)
    if sum/totalSum >= threshold:
        break
print("Number of dimensions left after dimensionality reduction: ", dimensions)
```

```python
reducedData = Z @ eigenVectors
reducedData = reducedData[:,:dimensions]
```

```python
reducedDf = pd.DataFrame(reducedData)
reducedDf.columns = ['feat_'+ str(i) for i in range(reducedData.shape[1])]
```
